Решение реальных задач математического программирования включает в себя следующие этапы:


1.   **Формализация исходной проблемы**. На этом этапе осуществляется исследование предметной области, определение целевой функции, построение системы ограничений; 
2.   **Построение математической модели**. Перефод формализованной задачи, описание которой получено на предыдущем этапе, на чёткий язык математических отношений;
3.   **Решение модели**. Нахождение оптимального управления при помощи алгоритмов математической оптимизации; 
4.   **Проверка адекватности модели**. Проверка решения на предмет наличия противоречий, некорректных элементов решения.
